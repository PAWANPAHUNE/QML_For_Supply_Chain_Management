```python
import pandas as pd
import numpy as np
from sklearn.utils import resample
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score, cross_val_predict
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data
train_df = pd.read_csv("Training_BOP_preprocessed.csv")
test_df = pd.read_csv("Testing_BOP_preprocessed.csv")

# 2. Balance the training data (1:1 with 5000 samples per class)
def balance_data(df, target_col, n_samples_per_class):
    class_0 = df[df[target_col] == 0]
    class_1 = df[df[target_col] == 1]

    class_0_sampled = resample(class_0, replace=False, n_samples=n_samples_per_class, random_state=42)
    class_1_sampled = resample(class_1, replace=False, n_samples=n_samples_per_class, random_state=42)

    balanced_df = pd.concat([class_0_sampled, class_1_sampled]).sample(frac=1, random_state=42).reset_index(drop=True)
    return balanced_df

train_balanced = balance_data(train_df, 'went_on_backorder', 5000)
test_balanced = balance_data(test_df, 'went_on_backorder', 2000)

X_train = train_balanced.drop(columns=['went_on_backorder'])
y_train = train_balanced['went_on_backorder']

X_test = test_balanced.drop(columns=['went_on_backorder'])
y_test = test_balanced['went_on_backorder']

print(f"Balanced Train Shape: {X_train.shape}, Balanced Test Shape: {X_test.shape}")

# 3. Hyperparameter tuning with RandomizedSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2],
    'reg_alpha': [0, 0.01, 0.1],
    'reg_lambda': [1, 1.5, 2.0]
}

xgb_clf = xgb.XGBClassifier(objective='binary:logistic', use_label_encoder=False, eval_metric='logloss', random_state=42)

search = RandomizedSearchCV(
    xgb_clf, param_distributions=param_grid,
    n_iter=25, scoring='accuracy', cv=5, verbose=1, random_state=42, n_jobs=-1
)

search.fit(X_train, y_train)
best_model = search.best_estimator_

# 4. Evaluate full feature model
print("\n--- Full Feature Model ---")

# Training Accuracy
train_pred = best_model.predict(X_train)
train_acc = accuracy_score(y_train, train_pred)
print(f"Training Accuracy: {train_acc:.4f}")
print("Confusion Matrix (Train):\n", confusion_matrix(y_train, train_pred))

# Testing Accuracy
test_pred = best_model.predict(X_test)
test_acc = accuracy_score(y_test, test_pred)
print(f"Testing Accuracy: {test_acc:.4f}")
print("Confusion Matrix (Test):\n", confusion_matrix(y_test, test_pred))

# 5-fold Cross Validation Accuracy
cv_scores = cross_val_score(best_model, X_train, y_train, cv=5, scoring='accuracy')
print(f"5-Fold CV Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")

# Classification Report
print("\nClassification Report (Test Data):")
print(classification_report(y_test, test_pred))

# 5. Feature importance (Gain)
booster = best_model.get_booster()
importance = booster.get_score(importance_type='gain')
sorted_importance = sorted(importance.items(), key=lambda x: x[1], reverse=True)
top_3_features = [k for k, v in sorted_importance[:3]]
print(f"\nTop 3 features based on Gain: {top_3_features}")

# Optional: Visualize top 10 feature importance
plt.figure(figsize=(10, 5))
xgb.plot_importance(best_model, max_num_features=10, importance_type='gain')
plt.title("Top 10 Features (Gain)")
plt.tight_layout()
plt.show()

# 6. Retrain using top 3 features
print("\n--- Top 3 Feature Model ---")
X_train_top3 = X_train[top_3_features]
X_test_top3 = X_test[top_3_features]

best_model_top3 = xgb.XGBClassifier(**search.best_params_, use_label_encoder=False, eval_metric='logloss')
best_model_top3.fit(X_train_top3, y_train)

# Training Accuracy
train_pred_top3 = best_model_top3.predict(X_train_top3)
train_acc_top3 = accuracy_score(y_train, train_pred_top3)
print(f"Training Accuracy: {train_acc_top3:.4f}")
print("Confusion Matrix (Train):\n", confusion_matrix(y_train, train_pred_top3))

# Testing Accuracy
test_pred_top3 = best_model_top3.predict(X_test_top3)
test_acc_top3 = accuracy_score(y_test, test_pred_top3)
print(f"Testing Accuracy: {test_acc_top3:.4f}")
print("Confusion Matrix (Test):\n", confusion_matrix(y_test, test_pred_top3))

# 5-fold Cross Validation Accuracy
cv_scores_top3 = cross_val_score(best_model_top3, X_train_top3, y_train, cv=5, scoring='accuracy')
print(f"5-Fold CV Accuracy: {cv_scores_top3.mean():.4f} (+/- {cv_scores_top3.std():.4f})")

# Classification Report
print("\nClassification Report (Test Data):")
print(classification_report(y_test, test_pred_top3))

# Final dataset shapes
print("\nFinal training data shape (top 3 features):", X_train_top3.shape)
print("Final testing data shape (top 3 features):", X_test_top3.shape)
```

```python
# Save the training and testing datasets with top 3 features
X_train_top3['went_on_backorder'] = y_train.values
X_test_top3['went_on_backorder'] = y_test.values

# Save to CSV
X_train_top3.to_csv("Training_Top3Features.csv", index=False)
X_test_top3.to_csv("Testing_Top3Features.csv", index=False)

print("Top 3 feature training and testing datasets saved successfully.")
```
